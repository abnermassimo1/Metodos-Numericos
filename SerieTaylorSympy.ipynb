Usaremos sympy para obtener la serie truncada de Taylor de: $$f (x) = ln(x)$$ de cuarto orden con x0 = 1.

Graficaremos f(x) junto con el polinomio de Taylor.

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
x0=1
#n=5 porque incluye hasta el término de cuarto orden
n=5
```

```python
# Definir la variable simbólica y la función
x = sp.symbols('x')
f = sp.ln(x)
```

```python
# Calcular la serie de Taylor de cuarto orden alrededor de x0 = 1
taylor_series = sp.series(f,x,x0,n).removeO()
```

```python
# Convertir la serie de Taylor a una función numérica
taylor_func = sp.lambdify(x, taylor_series, 'numpy')
```

```python
# Convertir la función original a una función numérica
f_func = sp.lambdify(x, f, 'numpy')

# Crear un array de valores x para la gráfica
x_values = np.linspace(0.2, 2, 400)

# Calcular los valores de y para ambas funciones
y_f = f_func(x_values)
y_taylor = taylor_func(x_values)

# Graficar ambas funciones
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_f, label='ln(x)', color='blue')
plt.plot(x_values, y_taylor, label='Aproximación de Taylor (4to orden)', color='red', linestyle='--')

# Añadir título y etiquetas
plt.title('Comparación de ln(x) y su Aproximación de Taylor de 4to orden')
plt.xlabel('x')
plt.ylabel('y')

# Añadir una leyenda
plt.legend()

# Mostrar la cuadrícula
plt.grid(True)

# Mostrar la gráfica
plt.show()

# Imprimir la serie de Taylor
sp.pprint(taylor_series)
```
